In this project, a KPI dashboard was created using the Restaurant Business Rankings 2020 dataset from Kaggle. The objective was to help managers visualize and monitor the business performance of the restaurant sector through key metrics such as annual sales, unit growth, and average sales per location.

Python, pandas, and matplotlib were used to clean the data and generate visual reports showing trends and rankings.

```python
# Importer les bibliothèques nécessaires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Charger les données (ajuste le chemin si besoin)
df = pd.read_csv('/content/Future50.csv')

# Afficher les premières lignes pour vérifier
print(df.head())
```

```python
# Nettoyage des données
# Supprimer les lignes avec des valeurs manquantes
df = df.dropna()

# Afficher les top 10 restaurants par ventes annuelles
top_10_sales = df[['Restaurant', 'Sales']].sort_values(by='Sales', ascending=False).head(10)
```

```python
# Visualisation : Top 10 Restaurants par Chiffre d'Affaires
plt.figure(figsize=(12,6))
sns.barplot(data=top_10_sales, x='Sales', y='Restaurant', palette='viridis')
plt.title('Top 10 Restaurants by Annual Sales')
plt.xlabel('Sales (in million USD)')
plt.ylabel('Restaurant')
plt.grid(axis='x')
plt.show()
```

```python
# Répartition du nombre d'unités par restaurant
plt.figure(figsize=(12,6))
sns.histplot(df['Units'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Number of Units')
plt.xlabel('Number of Units')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Chiffre d'affaires moyen par unité
df['Sales_per_Unit'] = df['Sales'] / df['Units']
average_sales_per_unit = df['Sales_per_Unit'].mean()
print(f'Average Sales per Unit: {average_sales_per_unit:.2f} million USD')
```
